# Feature Selection Advanced House Price Prediction

The main of this project is to predict the house price based on various features which we will discuss as we go aheaad 

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

##for feature selection

from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

#to visualise all the columns in the dataframe
pd.pandas.set_option('display.max_columns',None)
```

```python
ds=pd.read_csv('X_train.csv')
```

```python
ds.head()
```

```python
## capture the dependent feature
y_train=ds[['SalePrice']]
```

```python
## drop dependent feature from dataset
X_train=ds.drop(['Id','SalePrice'],axis=1)
```

```python
###Apply feature Selection
#first ,i specify the Lasso Regression model, and i
#select a suitable alpha(equivalent of penality)
# the bigger the alpha the less features that will be selected

#then i use selectfrommodel object froom sklearn , which
# will select thr feature which coefficient are non zero

feature_sel_model=SelectFromModel(Lasso(alpha=0.005,random_state=0))  
feature_sel_model.fit(X_train,y_train)

```

```python
feature_sel_model.get_support()
```

```python
# Let's print the no of total and selected features
#this is how we can make a list of the selected features
selected_feat=X_train.columns[(feature_sel_model.get_support())]

#let's print some stats
print('total feature: {}'.format((X_train.shape[1])))
print('selected features:{}'.format(len(selected_feat)))
print('feature with coefficients shrank to zero: {}'.format(
    np.sum(feature_sel_model.estimator_.coef_==0)))
```
